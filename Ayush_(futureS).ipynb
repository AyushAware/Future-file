```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

```

```python

data = pd.read_csv(io.BytesIO(uploaded['FUTSTK_CIPLA_01-Jan-2020_TO_26-Mar-2020.csv']))
```

```python
st = pd.DataFrame(data ,index=None, columns=['Settle Price'] )
st.to_numpy()

strike_price  = 480.15
```

```python
def call_payoff(st, strike_price):
  return np.where(st > strike_price, st - strike_price, st - strike_price)

payoff_long_future = call_payoff(st, strike_price)

# Plot
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False) # Top border removed
ax.spines['right'].set_visible(False) # Right border removed
ax.spines['bottom'].set_position('zero') # Sets the X-axis in the center
ax.plot(st,payoff_long_future,label='Long Call',color='g')
plt.xlabel('Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
def short_payoff(st, strike_price):
    return np.where(strike_price > st, strike_price - st, strike_price - st)

payoff_long_future = short_payoff (st, strike_price)
# Plot
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False) # Top border removed
ax.spines['right'].set_visible(False) # Right border removed
ax.spines['bottom'].set_position('zero') # Sets the X-axis in the center
ax.plot(st,payoff_long_future,label='Short Call',color='g')
plt.xlabel('Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```
